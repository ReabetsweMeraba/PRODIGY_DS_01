## Prepare Environment

```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
```

## Load DataSet

```python
file_path = 'dataset.csv'
data = pd.read_csv(file_path, skiprows=5, encoding='utf-8-sig')
```

##  Extract the first 30 rows and focus on the most recent year

```python
recent_year = data.columns[-2]  
chart_data = data.iloc[:30, [0, data.columns.get_loc(recent_year)]]
chart_data.columns = ['Country/Region', 'Population']
chart_data['Population'] = pd.to_numeric(chart_data['Population'], errors='coerce')
```

## Plot Bar Graph 

```python
plt.figure(figsize=(15, 8))
plt.bar(chart_data['Country/Region'], chart_data['Population'], color='skyblue')
plt.xticks(rotation=45, ha='right')
plt.title(f"Population by Country/Region ({recent_year})", fontsize=16)
plt.ylabel("Population", fontsize=12)
plt.xlabel("Country/Region", fontsize=12)
plt.tight_layout()
plt.show()
```
